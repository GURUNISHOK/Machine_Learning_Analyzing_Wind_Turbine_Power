```python
Hours_of_data_predicted = [12,24,48,168,672]
Root_Mean_Squared_Error = [16.036,31.495,239.609,58.774,56.040]
```

```python
def Average(lst): 
    return sum(lst) / len(lst) 
Root_Mean_Squared_Error = [16.036,31.495,239.609,58.774,56.040] 
average = Average(Root_Mean_Squared_Error) 
print("Average RMSE =", round(average, 2)) 
```

```python

import matplotlib.pyplot as plt
plt.xlabel('Hours of predicted data')
plt.ylabel('Root Mean Square Error')
plt.plot(Hours_of_data_predicted,Root_Mean_Squared_Error)
plt.show()
```
